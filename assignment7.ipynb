```python
with open('rosalind_gc.txt') as f:
    lines = [line.strip() for line in f.readlines()]
    
idx = 0
dna_strs = {}
max_gc = 0
max_gc_str = ""

while idx < len(lines):
    if lines[idx][0] == '>':
        label = lines[idx].split('>')[1]
        dna_str = ''
        idx += 1

        while idx < len(lines) and lines[idx][0] != '>':
            dna_str += lines[idx]
            idx += 1
        
        gc = (dna_str.count('G') + dna_str.count('C')) / len(dna_str) * 100
        if gc > max_gc:
            max_gc = gc
            max_gc_str = label
            
print(max_gc_str)
print(max_gc)
```

```python
with open('rosalind_grph-4.txt') as f:
    lines = [line.strip() for line in f.readlines()]

dna_strs = []
idx = 0
while idx < len(lines):
    if lines[idx][0] == '>':
        label = lines[idx].split('>')[1]
        dna_str = ''
        idx += 1

        while idx < len(lines) and lines[idx][0] != '>':
            dna_str += lines[idx]
            idx += 1
        
    dna_strs.append((label, dna_str))

my_list = []
for i in range(len(dna_strs)):
    for j in range(len(dna_strs)):
        if i == j: continue
        if dna_strs[i][1][-3:] == dna_strs[j][1][:3]: 
            print(dna_strs[i][0]+" "+dna_strs[j][0])
            my_list.append((dna_strs[i][0], dna_strs[j][0]))
```
